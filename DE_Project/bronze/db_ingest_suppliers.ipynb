Notebook: db_ingest_suppliers
Layer: Ingestion → Bronze
Source: Database (CSV simulation from ADLS)
Description:
- Reads raw suppliers data
- Adds ingestion metadata
- Writes to Bronze Delta table
Architecture:
Local Upload (simulating ADLS) → Spark → Bronze

Step 1: Read raw data from source

```python
df = spark.table("bronze.suppliers")
```

 Step 2: Data inspection

```python
df.count()
df.printSchema()
display(df)
```

Step 3: Add ingestion metadata

```python
from pyspark.sql import functions as F
df = df.withColumn("ingestion_timestamp", F.current_timestamp()) \
       .withColumn("source", F.lit("database"))
```

Step 4: Write to Bronze Delta table


```python
df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze.suppliers")
```

 Step 5: Validate output

```python
display(spark.table("bronze.suppliers"))
```
