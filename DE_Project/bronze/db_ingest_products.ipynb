 Notebook: db_ingest_products
 Layer: Ingestion → Bronze
 Source: Database (CSV simulation from ADLS)# Description:
 - Reads raw products data
 - Adds ingestion metadata
 - Writes to Bronze Delta table

 Architecture:
 Local Upload (simulating ADLS) → Spark → Bronze

 Step 1: Read raw data from source

```python
df = spark.table("bronze.products")
```

Step 2: Data inspection

```python
df.count()
df.printSchema()
display(df)
```

Step 3: Add ingestion metadata

```python
from pyspark.sql.functions import current_timestamp, lit

df = df.withColumn("ingestion_timestamp", current_timestamp()) \
       .withColumn("source", lit("database"))
```

Step 4: Write to Bronze Delta table

```python
df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze.products")
```

 Step 5: Validate output

```python
display(spark.table("bronze.products"))
```
