 Notebook: db_ingest_stores
 Layer: Ingestion → Bronze
 Source: Database (CSV simulation from ADLS)
 Description:
 - Reads raw stores data
 - Adds ingestion metadata
 - Writes to Bronze Delta table

 Architecture:
 Local Upload (simulating ADLS) → Spark → Bronze

```python
df = spark.table("bronze.stores")
```

```python
df.count()
df.printSchema()
display(df)
```

```python
from pyspark.sql import functions as F

df = df.withColumn("ingestion_timestamp", F.current_timestamp()) \
       .withColumn("source", F.lit("database"))
```

```python
df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("bronze.stores")
```
